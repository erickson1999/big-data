```python
# Importa la biblioteca pandas y la asigna a la abreviatura 'pd'
import pandas as pd
```

```python
data = pd.read_csv("https://docs.google.com/spreadsheets/d/e/2PACX-1vSFi9S08_nMnWIjOUeAm9pKkDmkfU2jBZjpqwKVv3-GU8STQJnTDMmuSCGCD4JynygoZg3QKZKce_1i/pub?gid=176273720&single=true&output=csv")
```

```python
#SUMA TOTAL DE INGRESOS
# Calcula la suma de la columna "ingresos" en el DataFrame 'data' utilizando el método sum() de Pandas
suma_ingresos = data["ingresos"].sum()
# Imprime un mensaje indicando que se mostrará la suma total
print("la suma total es: ")
# Imprime la suma total de ingresos que se calculó anteriormente
print(suma_ingresos)
```

```python
#PROMEDIO DE EDAD
# Calcula el promedio de la columna "edad" en el DataFrame 'data' utilizando el método mean() de Pandas
# Luego, redondea el resultado al número entero más cercano utilizando el método round()
prom_edad = data["edad"].mean().round()
# Imprime un mensaje indicando que se mostrará el promedio de edad
print("el promedio de edad es: ")
# No se está imprimiendo explícitamente el promedio de edad, solo se está evaluando
# Sin embargo, el promedio de edad se mostrará en la salida, ya que la variable se evalúa sin imprimirse directamente
prom_edad
```

```python
#PERSONAS DE AREQUIPA
# Selecciona las filas del DataFrame 'data' donde el valor en la columna "ciudad" sea igual a "Arequipa"
personas_arequipa = data.loc[data["ciudad"]=="Arequipa"]
# Imprime el DataFrame 'personas_arequipa', que contiene solo las filas donde la ciudad es "Arequipa"
personas_arequipa
```

```python
#promedio por ciudad
# Calcula el promedio de ingresos por ciudad agrupando el DataFrame 'data' por la columna "ciudad"
# y luego aplicando el método mean() a la columna "ingresos"
prom_in_ciudad = data.groupby("ciudad")["ingresos"].mean()
# Convierte los valores resultantes a tipo entero utilizando el método astype(int)
prom_in_ciudad = prom_in_ciudad.astype(int)
# Imprime el promedio de ingresos por ciudad
prom_in_ciudad
```

```python
#OBTENEMOS LA PERSONA MÁS JOVE
# Selecciona la primera fila donde la edad es igual a la edad mínima en el DataFrame 'data'
joven = data.loc[data["edad"]==data["edad"].min()]
# Imprime la persona más joven
joven
```

```python
#obtenemos la persona más mayor
# Selecciona la primera fila donde la edad es igual a la edad máxima en el DataFrame 'data'
mayor = data.loc[data['edad']==data["edad"].max()]
# Almacena esta fila en la variable 'mayor' y la imprime
joven
```

```python
#OBTENEMOS EL NUMERO DE PERSONAS POR CIUDAD
# Calcula el número de personas por ciudad agrupando el DataFrame 'data' por la columna "ciudad"
# y contando el número de valores no nulos en la columna "nombre" utilizando el método count()
personas_por_ciudad = data.groupby("ciudad")["nombre"].count()

# Imprime el número de personas por ciudad
personas_por_ciudad
```

```python
#ORDEN POR EDAD DESCENDENTE
# Ordena el DataFrame 'data' en orden descendente según los valores de la columna "edad"
orden_edad_descendente = data.sort_values(by="edad", ascending=False)
# Imprime el DataFrame resultante, que contiene filas ordenadas por edad en orden descendente
orden_edad_descendente
```

```python
# PROMEDIO DE INGRESOS DE MAYORES DE 30 AÑOS
# Selecciona las filas del DataFrame 'data' donde la edad es mayor que 30
mayores_30 = data[data["edad"] > 30]
# Calcula el promedio de ingresos de las personas mayores de 30 años
promedio_ingresos_mayores_30 = mayores_30["ingresos"].mean()
# Imprime el promedio de ingresos redondeado de las personas mayores de 30 años
print("El promedio de ingresos de mayores de 30 años es:", round(promedio_ingresos_mayores_30))
```

```python
#Ingresos totales por ciudades
# Calcula la suma de ingresos por ciudad agrupando el DataFrame 'data' por la columna "ciudad"
# y sumando los valores de la columna "ingresos" utilizando el método sum()
total_ciudad = data.groupby("ciudad")["ingresos"].sum()
# Imprime el total de ingresos por ciudad
total_ciudad
```

```python
# PERSONA CON EL INGRESO MAS ALTO
# Selecciona la(s) fila(s) donde el ingreso es igual al ingreso máximo en el DataFrame 'data'
persona_ingreso_mas_alto = data[data["ingresos"] == data["ingresos"].max()]
# Imprime un mensaje
print("Persona con el ingreso mas alto:\n")
# Muestra la(s) fila(s) correspondiente(s)
persona_ingreso_mas_alto
```

```python
# HISTOGRAMA DE EDADES
# Crea un histograma de la columna "edad" en el DataFrame 'data'
data["edad"].hist()
# Importa la biblioteca Matplotlib para visualización de datos
import matplotlib.pyplot as plt
# Muestra el histograma
plt.show()
```
