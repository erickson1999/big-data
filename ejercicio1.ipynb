```python
import pandas as pd
```

```python
data = pd.read_csv("https://docs.google.com/spreadsheets/d/e/2PACX-1vSFi9S08_nMnWIjOUeAm9pKkDmkfU2jBZjpqwKVv3-GU8STQJnTDMmuSCGCD4JynygoZg3QKZKce_1i/pub?gid=176273720&single=true&output=csv")
```

```python
print(data)
```

```python
#suma total de ingresos
suma_ingresos = data["ingresos"].sum()
print("la suma total es: ", suma_ingresos)
```

```python
#promedio de edad
prom_edad = data["edad"].mean()
print("el promedio de edad es: ",round(prom_edad))
```

```python
#personas en Arequipa
personas_arequipa = data.loc[data["ciudad"]=="Arequipa"]
print(personas_arequipa)
```

```python
#promedio por ciudad
prom_in_ciudad = data.groupby("ciudad")["ingresos"].mean().astype(int)
prom_in_ciudad
```

```python
#obtenemos la persona más joven
joven = data.loc[data["edad"]==data["edad"].min()].values[0]
joven
```

```python
#obtenemos la persona más mayor
mayor = data.loc[data['edad']==data["edad"].max()].values[0]
mayor
```

```python
#obtenemos el número de personas por ciudad - Modo 1
personas_por_ciudad = data.groupby("ciudad")["nombre"].count()
personas_arequipa
```

```python
#obtenemos el número de personas por ciudad - Modo 2
personas_ciudad = data["ciudad"].value_counts()
personas_ciudad
```

```python
#orden por edad descendente
orden_edad_descendente = data.sort_values(by="edad",ascending=False)
orden_edad_descendente
```
